```python
print("\nTop 5 states by total cancer rate:")
print(df.nlargest(5, 'Total.Rate')[['State', 'Total.Rate']])
print("\nBottom 5 states by total cancer rate:")
print(df.nsmallest(5, 'Total.Rate')[['State', 'Total.Rate']])
```

```python
age_columns = ['Rates.Age.< 18', 'Rates.Age.18-45', 'Rates.Age.45-64', 'Rates.Age.> 64']
age_data = df[age_columns]
plt.figure(figsize=(10, 6))
sns.boxplot(data=age_data)
plt.title('Distribution of Cancer Rates by Age Group')
plt.ylabel('Cancer Rate per 100,000')
plt.xticks(range(4), ['<18', '18-45', '45-64', '>64'])
plt.show()
```

```python
gender_columns = ['Rates.Age and Sex.Female.18 - 45', 'Rates.Age and Sex.Male.18 - 45',
                  'Rates.Age and Sex.Female.45 - 64', 'Rates.Age and Sex.Male.45 - 64',
                  'Rates.Age and Sex.Female.> 64', 'Rates.Age and Sex.Male.> 64']
gender_data = df[gender_columns]
plt.figure(figsize=(12, 6))
sns.boxplot(data=gender_data)
plt.title('Distribution of Cancer Rates by Age Group and Gender')
plt.ylabel('Cancer Rate per 100,000')
plt.xticks(range(6), ['F 18-45', 'M 18-45', 'F 45-64', 'M 45-64', 'F >64', 'M >64'])
plt.show()
```

```python
race_columns = ['Rates.Race.White', 'Rates.Race.Black', 'Rates.Race.Asian', 'Rates.Race.Indigenous', 'Rates.Race.Hispanic']
race_data = df[race_columns]
plt.figure(figsize=(10, 6))
sns.boxplot(data=race_data)
plt.title('Distribution of Cancer Rates by Race')
plt.ylabel('Cancer Rate per 100,000')
plt.xticks(range(5), ['White', 'Black', 'Asian', 'Indigenous', 'Hispanic'])
plt.show()
```

```python
cancer_columns = ['Types.Breast.Total', 'Types.Colorectal.Total', 'Types.Lung.Total']
cancer_data = df[cancer_columns]
plt.figure(figsize=(10, 6))
sns.boxplot(data=cancer_data)
plt.title('Distribution of Cancer Rates by Type')
plt.ylabel('Cancer Rate per 100,000')
plt.xticks(range(3), ['Breast', 'Colorectal', 'Lung'])
plt.show()
```

```python
correlation_columns = ['Total.Rate'] + age_columns + cancer_columns
correlation_data = df[correlation_columns]
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Cancer Rates')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
df.plot(kind='bar', x='State', y='Total.Rate')
plt.title('Total Cancer Rate by State')
plt.xlabel('State')
plt.ylabel('Cancer Rate per 100,000')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
key_metrics = ['Total.Rate', 'Types.Breast.Total', 'Types.Colorectal.Total', 'Types.Lung.Total']
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
for i, metric in enumerate(key_metrics):
    sns.histplot(df[metric], ax=axes[i//2, i%2], kde=True)
    axes[i//2, i%2].set_title(f'Distribution of {metric}')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[key_metrics])
plt.suptitle('Scatterplot Matrix of Cancer Types', y=1.02)
plt.show()
```

```python
race_columns = ['Rates.Race.White', 'Rates.Race.Black', 'Rates.Race.Asian', 'Rates.Race.Indigenous', 'Rates.Race.Hispanic']
df['Max_Racial_Disparity'] = df[race_columns].max(axis=1) - df[race_columns].min(axis=1)
plt.figure(figsize=(12, 6))
plt.bar(df['State'], df['Max_Racial_Disparity'])
plt.xticks(rotation=90)
plt.ylabel('Maximum Racial Rate Difference')
plt.title('Maximum Racial Disparity in Cancer Rates by State')
plt.show()
```

```python
df['Female_Percentage'] = (df['Rates.Age and Sex.Female.< 18'] + df['Rates.Age and Sex.Female.18 - 45'] +
                           df['Rates.Age and Sex.Female.45 - 64'] + df['Rates.Age and Sex.Female.> 64']) / df['Total.Rate'] * 100
plt.figure(figsize=(10, 6))
plt.scatter(df['Female_Percentage'], df['Types.Breast.Total'])
plt.xlabel('Female Percentage of Population')
plt.ylabel('Breast Cancer Rate')
plt.title('Breast Cancer Rate vs Female Population Percentage')
for i, state in enumerate(df['State']):
    plt.annotate(state, (df['Female_Percentage'][i], df['Types.Breast.Total'][i]))
plt.show()
```
